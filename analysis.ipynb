```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path_beginner = r"C:\Users\Startklar\Documents\GitHub\DeutscheApp\wortschatz_Beginner.csv"

vocab_df = pd.read_csv(file_path_beginner, encoding='iso-8859-1', sep=",")

vocab_df.replace({"Ã¼": "ü", "Ã¤": "ä", "Ã¶": "ö", "ÃŸ": "ß"}, regex=True, inplace=True)
vocab_df.columns = vocab_df.columns.str.strip()
```

```python
vocab_df.head()
```

```python
article_counts = vocab_df['Artikel'].value_counts()
article_percentages = (article_counts / len(vocab_df)) * 100
article_percentages
```

```python
plt.figure(figsize=(8, 8))
plt.pie(article_percentages, labels=article_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Articles")
plt.show()
```
