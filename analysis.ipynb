```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path_beginner = r"C:\Users\Startklar\Documents\GitHub\DeutscheApp\wortschatz_Beginner.csv"
file_path_intermediate = r"C:\Users\Startklar\Documents\GitHub\DeutscheApp\wortschatz _Intermediate.csv"
file_path_advanced = r"C:\Users\Startklar\Documents\GitHub\DeutscheApp\wortschatz_Advanced.csv"

beg_df = pd.read_csv(file_path_beginner, encoding='iso-8859-1', sep=",")
int_df = pd.read_csv(file_path_intermediate, encoding='iso-8859-1', sep=",")
adv_df = pd.read_csv(file_path_advanced, encoding='iso-8859-1', sep=",")

beg_df.replace({"Ã¼": "ü", "Ã¤": "ä", "Ã¶": "ö", "ÃŸ": "ß"}, regex=True, inplace=True)
int_df.replace({"Ã¼": "ü", "Ã¤": "ä", "Ã¶": "ö", "ÃŸ": "ß"}, regex=True, inplace=True)
adv_df.replace({"Ã¼": "ü", "Ã¤": "ä", "Ã¶": "ö", "ÃŸ": "ß"}, regex=True, inplace=True)
beg_df.columns = beg_df.columns.str.strip()
int_df.columns = int_df.columns.str.strip()
adv_df.columns = adv_df.columns.str.strip()
```

```python
beg_df.head()
```

```python
beg_article = beg_df['Artikel'].value_counts()
int_article = int_df['Artikel'].value_counts()
adv_article = adv_df['Artikel'].value_counts()
beg_percentages = (beg_article / len(beg_df)) * 100
int_percentages = (int_article / len(int_df)) * 100
adv_percentages = (adv_article / len(adv_df)) * 100

beg_percentages
```

```python
beg_counts = beg_article.reindex(['der', 'das', 'die'], fill_value=0)
int_counts = int_article.reindex(['der', 'das', 'die'], fill_value=0)
adv_counts = adv_article.reindex(['der', 'das', 'die'], fill_value=0)

total_counts = beg_counts + int_counts + adv_counts
grand_total = total_counts.sum()
overall_percentages = (total_counts / grand_total) * 100
```

```python
print("Overall Percentages:")
for article, percentage in overall_percentages.items():
    print(f"{article}: {percentage:.1f}%")
```

```python
plt.figure(figsize=(8, 8))
plt.pie(beg_percentages, labels=beg_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Articles (Beginner)")
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(int_percentages, labels=int_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Articles (Intermediate)")
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(adv_percentages, labels=adv_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Articles (Advanced)")
plt.show()

# Plot overall pie chart
plt.figure(figsize=(8, 8))
plt.pie(overall_percentages, labels=overall_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title("Overall Proportion of Articles")
plt.show()
```
